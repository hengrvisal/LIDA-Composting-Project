```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("lida_tabletrimmed.csv")
```

Checking unique values for device_id

```python
df['device_id'].unique()
```

Removing device_id due to being a single value

```python
df.drop('device_id', axis=1, inplace=True)
```

```python
# Check for lid unique value
df['lid'].unique()
```

```python
# Changing the text values to integers
# 1 for open and 0 for closed
df['lid_open'] = df['lid'].map({'open': 1, 'closed': 0})
```

```python
df.drop('lid', axis=1, inplace=True)
```

```python
# Creating Average Active Temperature Column
active_temp_cols = [col for col in df.columns if col.startswith('temperature_active')]
if active_temp_cols:
    df['temperature_active_avg'] = df[active_temp_cols].mean(axis=1)
```

```python
# Creating Max Temperature and Min Temperature

df['max_active_temperature'] = df[['temperature_active1', 'temperature_active2', 
                               'temperature_active3', 'temperature_active4']].max(axis=1)

df['min_active_temperature'] = df[['temperature_active1', 'temperature_active2',
                               'temperature_active3', 'temperature_active4']].min(axis=1)
```

```python
# Creating Average Active Moisture Column
active_moisture_cols = [col for col in df.columns if col.startswith('moisture_active')]
if active_moisture_cols:
    df['moisture_active_avg'] = df[active_moisture_cols].mean(axis=1)
```

```python
df.head()
```

```python
df['time_stamp'] = pd.to_datetime(df['time_stamp'], dayfirst=True, errors='coerce')
df['hour'] = df['time_stamp'].dt.hour
df['dayofweek'] = df['time_stamp'].dt.dayofweek
df['month'] = df['time_stamp'].dt.month
```

```python
df.drop('time_stamp', axis=1, inplace=True)
```

```python
object_columns = df.select_dtypes(include=['object']).columns

# Convert all object columns to float
for col in object_columns:
    df[col] = df[col].astype(float)
```

```python
# Creating a column for the ratio of oxygen to co2
df['o2_co2_ratio'] = df['oxygen'] / df['co2']
```

```python
df.to_csv('lida_tabletrimmed_featured.csv', index=False)
```
