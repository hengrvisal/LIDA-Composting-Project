```python
df = pd.read_csv('lida_tabletrimmed.csv')
df.head()
```

```python
temp_cols = [
    'temperature_active1',
    'temperature_active2',
    'temperature_active3',
    'temperature_active4',
    'temperature_curing1',
    'temperature_curing2'
]

df['overheat'] = df[temp_cols].max(axis=1) >= 70

AERATION_MAX = 15
AGITATION_MAX = 15

df['aeration_duration'] = df['overheat'].apply(lambda x: AERATION_MAX if x else 0)
df['agitation_duration'] = df['overheat'].apply(lambda x: AGITATION_MAX if x else 0)

```
